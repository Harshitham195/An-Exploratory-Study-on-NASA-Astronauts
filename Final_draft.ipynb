## Exploring Astronauts' Military Branch Diversity.

```python
%%sql
select military_branch, count(*) as number
from astronauts
group by military_branch;
```

## Top Military Ranks Among Astronauts.


```python
%%sql
select military_rank, count(*) as number
from astronauts
group by military_rank
order by number desc
limit 5;
```

## Exploring Astronaut Demographics: Gender Insights.

```python
%%sql
select gender, count(*) as number
from astronauts
group by gender;
```

## Cosmic Scholars: Top Graduate Majors of Astronauts.

```python
%%sql
select Graduate_major, count(Graduate_major) as number
from astronauts
group by Graduate_major
order by number desc;
```

## Educational Odyssey: Astronauts and Their Degrees.

```python
%%sql
select count(*) as no_of_Astronauts,
count(undergraduate_major) as astronauts_udergraduate_degree,
count(graduate_major) as astronautes_graduate_degree
from astronauts;
```

## Astronaut Birthplaces: Launching from Earth's Diverse Cities.


```python
%%sql
SELECT 
    CASE 
        WHEN instr(birth_place, ',') > 0 THEN substr(birth_place, instr(birth_place, ',') + 2)
        ELSE birth_place
    END as state, 
    COUNT(*) as number
FROM astronauts
GROUP BY state
ORDER BY number DESC
LIMIT 5;
```

## Voyages Beyond: Astronauts' Space Flights and Spacewalks.
